**Q.1 What is hypothesis testing in statistics?**
- Hypothesis testing is a statistical inference that uses data from a sample to draw conclusions about a population parameter. In other words, it is a claim or statement about a population parameter that can be tested using statistical methods. In hypothesis testing we try to reject null hypothesis using sample data by assuming it is true. Example: average salary of I.T employees in a specific country is 50k.

**Q.2 What is the null hypothesis, and how does it differ from the alternative hypothesis?**
- Null hypothesis is the initial or default assumption made on population which we try to reject using sample data. It represents the idea that nothing has changed and there is no effect/difference. Alternate hypothesis is opposite of null hypothesis. Like if the null hypothesis indicates that there is no difference between two or more groups whereas alternate hypothesis indicate that there is a difference between two or more groups.
Example:

  - Null hypothesis: The mediciene has no effect

  - Alternate hypothesis: The mediciene has effect

**Q.3 What is the significance level in hypothesis testing, and why is it important?**
- It is the threshold value users set before testing to decide how much risk they are willing to take in making a type 1 error. It is important because it represent the probability of rejecting a true null hypothesis in a statistical test. It is denoted by alpha. Example if alpha = 5% then we can accept 5% error in hypothesis testing.

**Q.4 What does a P-value represent in hypothesis testing?**
- P-value is a probability value, calculated from a statistical table (corressponding to z-score of a sample). It is used to decide whether we have to reject or accept null hypothesis. The smaller the p-value, the stronger the evidence will be there against null hypothesis. If p-value <= alpha(significance level) then reject null hypothesis and if p-value > alpha then we failed to reject null hypothsis.

**Q.5 How do you interpret the P-value in hypothesis testing?**
- The interpretation of p-value is crucial for deciding to accept or reject null hypothesis. P-value is calculated based on sample data and hypothesis testing from a statistical table. For interpretation we compare p-value with significance level. If p <= alpha, we reject null hypothesis and the results are statistical significant whereas if p-value > aplha then we accept null hypothesis which indicate that alternate hypothesis wasn't strong enough and the results are not statistical significant. P-value doesn't provide the probability that the null hypothesis is true but it tells us wheather the sample data / observed data is significant with null hypothesis or not.

**Q.6 What are Type 1 and Type 2 errors in hypothesis testing?**
- Both type-1 and type-2 errors are mistakes which occurs in hypothesis testing.
  - Type-1 error: It occurs when we reject null hypothesis when it is actually true i.e. we conclude that there is difference or effect but actually there isn't. Example: if we tell that a mediciene is effective in treating a disease but it actually have no effect.
  - Type-2 error: It occurs when we fail to reject null hypothesis when it is actually false i.e. we conclude that there is no difference but in reality there is a difference or effect. Example: a guilty person may not be convicted.

**Q.7 What is the difference between a one-tailed and a two-tailed test in hypothesis testing?**
- A one-tailed test is a uni-directional hypothesis where rejection region lies only on one side of the distribution. It is used to test that whether a parameter is greater than or less than a certain critical value. It of two types right tailed and left tailed. In right tailed critical region is on right side of the sampling distribution and in left tailed it is on left side of distribution. In contrast, two tailed test is bi-directional hypothesis where rejection/critical region lies on both sides of the graph. It is used to test when we have to check whether a parameter is equal to a certain value or not. This is used to test the effect in both the directions.

**Q.8 What is the Z-test, and when is it used in hypothesis testing?**
- z-test is a statistical test used in hypothesis testing to compare the population parameter with the sample parameter assuming that the data follows normal distribution. Like whether the popuation mean is equal to sample mean or not. Z-test is used when the sample size is greater than or equal to 30, standard deviation of population is known and the data is normally distributed. It is mainly used to check the population mean, differences between two means, population proportion and difference between two proportions.

**Q.9 How do you calculate the Z-score, and what does it represent in hypothesis testing?**
- Z-score is used to calculate how many standard deviations a sample statistic is from the imagined population parameter under the null hypothesis. Z-score is calculated using the following formula:
$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
$$
  - $\mu$ is a population mean
  - $\sigma$ is standard deviation
  - n is sample size
  - $\bar{X}$ is sample mean

In hypothesis testing it indicates how far our assumption is from the actual value assuming that the null hypothesis is true. If z-score is large it mean sample mean is far from hypothesized population mean. So, the evidence is against the null hypothesis. Similarly, if z-score is small it mean sample mean and hypothesized mean are close to each other and hence we fail to reject null hypothesis.

**Q.10 What is the T-distribution, and when should it be used instead of the normal distribution?**
- T-distribution is a statistical distribution similar to z-distribution used to make inferences about population but it is used when either sample size < 30 or population standard deviation is unknown. It's graph is similar to normal distribution but due to extreme values it has heavier tails. So, as sample size increases graph follows normal distribution.
$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$
  - $\mu$ is a population mean
  - s is sample standard deviation
  - n is sample size
  - $\bar{X}$ is sample mean

**Q.11 What is the difference between a Z-test and a T-test?**
- The key differences in Z-test and T-test lie in their assumptions. Z-test are used when we have large sample size i.e. n>=30 and when population standard deviation is known, whereas t-test are mostly are helpful when either n < 30 (sample size small) or population standard deviarion is unknown.

**Q.12 What is the T-test, and how is it used in hypothesis testing?**
- T-test is a statistical test used to population mean or proportion with a certain value. It is similar to z-test but is used when either sample size < 30 or population standard deviation is unknown or both. It uses t-score to test whether to accept or reject null hypothesis.
$$
t = \frac{\bar{X} - \mu}{s / \sqrt{n}}
$$
  - $\mu$ is a population mean
  - s is sample standard deviation
  - n is sample size
  - $\bar{X}$ is sample mean

  Along with t-score we also calculate degree of freedom which is equal to n(sample size)-1. From t-table we calculate probability value by using t-score and degree of freedom. Now, if probability value <= significance value we reject null hypothesis else we accept null hypothesis.

**Q.13 What is the relationship between Z-test and T-test in hypothesis testing?**
- Both z-test and t-test are used in hypothesis testing to determine the difference between population and sample parameter. They both follow the same approach and formula to test the hypothesis about means and the differences lie only in their assumptions. Both are used to compare population mean to a certain value or two population mean of two different sample data.

**Q.14 What is a confidence interval, and how is it used to interpret statistical results?**
- A confidence interval is the mean of point estimate plus and minus the variation in that estimate. It is the range of values we expect our estimate to fall in considering the tolerance of error. Here, the point estimate is located at the middle of confidence interval. Confidence interval gives a range where a population parameter would lie based on estimation of sample parameter. Also, The wider the interval the more confidence will be there but with less precision. Example 95% confidence interval doesn't mean that 95% values lie in that interval but it means that if we repeat the sample 100 times, about 95 of those confidence interval will have population parameter.

**Q.15 What is the margin of error, and how does it affect the confidence interval?**
- Margin of error is the maximum expected error which we can tolerate between sample parameter and population parameter. We use margin of error to calculate confidence interval as due to margin of error we know the uncertainity we can handle in the experiment. Margin of error decreases with increase in sample size and leads to more precise confidence intervals. Similarly, margin of error increases with decrease in sample size and leads to less precise confidence intervals.
$$
\text{Margin of Error} = Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$
  - $Z_{\alpha/2}$ is z-score corressponding to alpha/2
  - $\sigma$ is standard deviation
  - n is sample size

**Q.16 How is Bayes' Theorem used in statistics, and what is its significance?**
- Bayes' theorm provides a way to update the probability of a hypothesis or condition being true after taking into account new evidence or information. It calculates the reversed probability of A given B by relating to probability of B given A and the initial probability of A and B.
$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$
  It helps to determine the conditional probability of an event based on prior knowledge and new evidence. It offers probabilistic interpretation of parameters unlike point estimates. Hence, it is used to make informed decisions.

**Q.17 What is the Chi-square distribution, and when is it used?**
- Chi-square distribution is a probability distribution that describes the distribution of a sum of square of k random variables. In other words, if we square the number of any sample, it will closely follow chi-square distribution. The shape of chi-square distribution depends upon the degree of freedom which is equal to sample size - 1. It is non-negative and right skewed distribution. The chi-square distributions are used to identify the probability of a value occuring and this probability value can be used to determine how closely observed values fit expected values and whether the variables are independent of each other or not.

**Q.18 What is the Chi-square goodness of fit test, and how is it applied?**
- Chi-square goodness of fit test is used to compare observed and expected categorical data. It is a non-parametric tests i.e. there are no assumptions made to test this and it only depends on degree of freedom. In goodness of fit test null hypothesis states that there is no difference between expected and observed values and alternate hypothesis states that expected value is not equal to observed value.
$$
\chi^2 = \sum_{i=1}^{n} \frac{(O_i - E_i)^2}{E_i}
$$

  As it is right skewed data so, if chi-sq < critical value(calculated from table using significance level and degree of freedom) we fail to reject null hypothesis. In contrast, if chi-sq > critical value we reject null hypothesis.

**Q.19 What is the F-distribution, and when is it used in hypothesis testing?**
- The f-distribution is a probability distribution that is useful in the context of comparing variances of two or more samples. It is right-skewed and takes only non-negative values. F-distribution shape depends on d1 and d2 i.e. degree's of freedom of both samples. If both d1 and d2 are greater than 30 then the f-distribution behaves like a normal distribution.

  It is widely used to compare the variances of two or more samples.

  f-statistics = $\frac{s1^2}{s2^2}$
  where, s1 and s2 are sample standard deviations.

**Q.20 What is an ANOVA test, and what are its assumptions?**
- ANOVA stands for analysis of variance. It is a statistical method used to compare the means of two or more than two groups. In this we use variance to determine if the mean value of three or more than three groups are same or different.
Assumptions of ANOVA are listed below:
  - the population from which sample are drawn should be normally distributed
  - the sample should be independent of each other
  - outliers should not be present
  - variance should be homogenous i.e. variance among the group must be approximately equal.

**Q.21 What are the different types of ANOVA tests?**
- There are particularly three types of ANOVA and each of them are listd below:
  - *ONE-WAY ANOVA:* It has one factor with atleast two levels and each level is independent of each other. Example: stress levels of employees in 3 departments.
  - *REPEATED MEASURES ANOVA:* It has one factor with atleast two levels but each level is dependent upon each other. Example: Number of hours studied on monday, tuesday, wednesday. They all depend on each other like if we studied more on monday then we will study less on next day as we already completed the work on previous day.
  - *FACTORIAL ANOVA:* In this ANOVA test there are two or more factors and each of them has atleast two levels. Levels can be either dependent, independent or both. It has a subtype known as two-way anova in which there is only two factors.

**Q.22 What is the F-test, and how does it relate to hypothesis testing?**
- F-test is primarily used to compare variances of two or more samples in the context of ANOVA and regression analysis.To compare the variances of two or more independent samples the following formula is used.

  f-statistics = $\frac{s1^2}{s2^2}$
  where, s1 and s2 are sample standard deviations.

In f-test null hypothesis states that both the sample variances are equal where as alternate hypothesis states that there is some difference between them. And after calucating critical value from table using degree of freedom of both samples, we compare them with f-statistics to reject or accept null hypothesis. If f-stat <= f-critical we fail to reject i.e. accept null hypothesis and if f-test > f-critical then we reject null hypothesis.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pdf(x,p):
  '''
  beroulli distribution has two parameters x -> random variable which is equal to 0 or 1,
  p means probability of success which lies between 0 and 1.
  '''
  if not (0<=p<=1):
    raise ValueError('Probability is wrong')
  if x not in [0,1]:
    raise ValueError('Random variable can only be 0 or 1')
  return (p**x)*((1-p)**(1-x))

print(bernoulli_pdf(x=0,p=0.4))
```

```python
#9. Write a Python function to calculate Z-scores from a dataset and plot them
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
  '''
  this function first calculate the z-scores and then plot the histogram of these z-scores
  '''
  #convert data into nparray
  #data=np.array(data)
  mean=np.mean(data)
  std_dev=np.std(data)
  if std_dev ==0:
    raise ValueError('standard deviation can not be 0 to calculate z-scores')
  z_score=(data-mean)/std_dev
  plt.hist(z_score,bins=10,alpha=0.6,edgecolor='black')
  plt.xlabel('Z-score')
  plt.ylabel('Frequency')
  plt.title('Graph of z-scores')
  plt.show()
  return z_score
data=np.random.randint(1,100,10)
print(calculate_z_scores(data))
```

```python
#1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results
import numpy as np
from statsmodels.stats.weightstats import ztest
population = np.random.rand(10000000)
pop_mean = np.mean(population)
#Generate sample from population
sample = np.random.choice(population,size=100)

#Hypothesis
# NuLL: sample mean = pop_mean, Alternate : sample mean != pop_mean
# z-test function uses sample std dev so no need to explicily mention that
zscore,pvalue=ztest(sample,value=pop_mean,alternative="two-sided")

print('Population mean is ',pop_mean)
print('Sample mean is ',np.mean(sample))
print('Z-score is ',zscore)
print('P-value is ',pvalue)
if pvalue <= 0.05:
  print('Reject null hypothesis')
else:
  print('Fail to reject null hypothesis')
```

```python
#4. Perform a two-tailed Z-test using Python and visualize the decision region on a plot
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from statsmodels.stats.weightstats import ztest
population = np.random.normal(100,20,size=100000000)
pop_mean = np.mean(population)
#Generate sample from population
sample = np.random.choice(population,size=1000)

#Hypothesis
# NuLL: sample mean = pop_mean, Alternate : sample mean != pop_mean
# z-test function uses sample std dev so no need to explicily mention that
zscore,pvalue=ztest(sample,value=pop_mean,alternative="two-sided")

#plot :
# make z-scores as x-axis
x = np.linspace(-10,10,1000)
#pdf of standard normal distribution with mean 0 and std dev 1
y = norm.pdf(x,0,1)
alpha = 0.05
z_critical = norm.ppf(1-alpha/2) # it is right critical value but normal is symmtric so it will be equal to left one also
plt.figure(figsize=(10, 5))
plt.plot(x,y)
# Shade rejection regions
plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.3, label='Rejection Region')
# Plot z-score
plt.axvline(zscore, color='blue', linestyle='--', label=f'Z-score = {zscore}')
plt.legend()
plt.show()
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
def visualization_of_errors(mu0,mu1,sigma,alpha):
  '''
  For Right-tailed test i am visualizing type-1 and type-2 error...
  Type-1 error: Reject H0 when H0 is true
  Type-2 error: Accept H0 when H1 is true
  '''
  x=np.linspace(min(mu0, mu1) - 4*sigma, max(mu0, mu1) + 4*sigma, 1000)
  #Make distribution
  null_dist=norm(mu0,sigma)
  alt_dist=norm(mu1,sigma)
  #Plot graphs of both
  plt.plot(x,null_dist.pdf(x),label='NULL DISTRIBUTION')
  plt.plot(x,alt_dist.pdf(x),label='ALTERNATE DISTRIBUTION')
  z=norm.ppf(1-alpha) # Because it is right tailed
  critical_value=mu0+z*sigma # opp of z-score formula
  # Type I error (α)
  # Here, we shade rejection region as error occur when H0 true but we reject
  plt.fill_between(x, 0, null_dist.pdf(x), where=(x >= critical_value),color='red', alpha=0.3, label='Type 1 Error')

  # Type II error (β)
  #Here, we shade acceptance region as error occur when H1 true but we fail to reject
  plt.fill_between(x, 0, alt_dist.pdf(x), where=(x < critical_value),color='orange', alpha=0.3, label='Type 2 Error')
  plt.title("TYPE-1 AND TYPE-2 ERROR")
  plt.legend()
  plt.show()
visualization_of_errors(400,409,10,0.05)
```

```python
#6.  Write a Python program to perform an independent T-test and interpret the results.
# In t-test n<30 or sigma is unknown
import numpy as np
from scipy import stats
population=np.random.randint(10,50,100)
sample=np.random.choice(population,size=20)
pop_mu=np.mean(population)
# Make Hypothesis
# H0: Mean = pop_mu
# H1: Mean != pop_mu
# In t-test we know only about sample and population mean
tscore,p_value=stats.ttest_1samp(sample,pop_mu)
if p_value<=0.05:
  print('REJECT NULL HYPOTHESIS(POPULATION MEAN != SAMPLE MEAN)')
else:
  print('Fail to reject NUll hypothesis(POPULATION MEAN == SAMPLE MEAN)')
```

```python
# 7. Perform a paired sample T-test using Python and visualize the comparison results
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

population=np.random.randint(10,50,100)
sample1=np.random.choice(population,size=20)
sample2=np.random.choice(population,size=20)
#H0: Mean of sample1 = Mean of sample 2
# H1: Mean of sample1 != Mean of sample 2
ttest,p_val=stats.ttest_rel(sample1,sample2)
if p_val<=0.05:
  print('REJECT NULL HYPOTHESIS')
else:
  print('Fail to reject NUll hypothesis')
plt.figure(figsize=(10, 6))
x = np.arange(len(sample1))
plt.bar(x,sample1,label='Sample_1',alpha=0.5)
plt.bar(x,sample2,label='Sample_2',alpha=0.5)
plt.xlabel('index')
plt.ylabel('values')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 8. Simulate data and perform both Z-test and T-test, then compare the results using Python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
from statsmodels.stats.weightstats import ztest

population=np.random.randint(10,50,10000)
sample=np.random.choice(population,size=40)
#hypothesis :
# H0: Population mean = sample mean
# H1: Population mean != sample mean
population_mean=np.mean(population)
# Z-test
zscore,p_val_z=ztest(sample,value=population_mean)
if p_val_z<=0.05:
  print('REJECT NULL HYPOTHESIS in z-test')
else:
  print('Fail to reject NUll hypothesis in z-test')

# T-test
tscore,p_val_t=stats.ttest_1samp(sample,population_mean)
if p_val_t<=0.05:
  print('REJECT NULL HYPOTHESIS in t-test')
else:
  print('Fail to reject NUll hypothesis in t-test')
print(f'In z-test: z-score = {zscore} and p-value = {p_val_z}')
print(f'In t-test: t-score = {tscore} and p-value = {p_val_t}')
```

```python
# 9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance
import numpy as np
import scipy.stats as stats
def conf_interval(sample,cl=0.95):
  ci=stats.t.interval(cl,df=len(sample)-1,loc=np.mean(sample),scale=np.std(sample, ddof=1) / np.sqrt(len(sample)))
  return ci
data=[10,30,40,20,40,50,20,11,34,45,57,86,25,16,29,64,29,46,52]
print(conf_interval(data))
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data
import numpy as np
import scipy.stats as stats

def margin_of_error(sample,cl=0.95):
  n=len(sample)
  df=n-1
  std_err = np.std(sample, ddof=1) / np.sqrt(n)
  t_score = stats.t.ppf((1 + cl) / 2, df)
  moe = t_score * std_err
  return moe
data=[10,30,40,20,40,50,20,11,34,45,57,86,25,16,29,64,29,46,52]
print(margin_of_error(data))
```

```python
#11. Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process
# In bayesian inference we know something prior and then using evidence we change the hypothesis according to the result
# Let take example of testing a certain disease
# Let 1% people has disease P(D)=0.01 this is prior information
# 95% test is accurate when person has disease P(test_poss|disease)
# 92% test is accurate when person don't have disease P(test_neg|no_disease)
def bayesian_inference(p_d,p_poss_d,p_neg_not_d):
  p_not_d= 1-p_d
  p_poss_not_d=1-p_neg_not_d
  p_poss=p_poss_d * p_d + p_poss_not_d * p_not_d
  # Posterior probability: P(Disease | Positive)
  # i.e. probability of having a disease when test is possitive is
  p_d_poss = (p_poss_d * p_d)/p_poss
  return p_d_poss
print(bayesian_inference(0.01,0.95,0.92))
```
